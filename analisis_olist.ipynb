## Delivery Time Analysis

In this step, we calculate the time it takes for orders to be delivered.

This metric is critical for understanding customer experience and logistics performance.

```python
# convert to datetime
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'])

# calculate delivery time in days
orders['delivery_time_days'] = (
    orders['order_delivered_customer_date'] - orders['order_purchase_timestamp']
).dt.days

orders[['delivery_time_days']].head()
```

## Average Delivery Time

In this step, we calculate the average number of days it takes for an order to be delivered.

This metric helps evaluate logistics efficiency and customer experience.

```python
orders['delivery_time_days'].mean()
```

## Business Insight - Average Delivery Time

The average delivery time is approximately 12 days.

This indicates the typical waiting time customers experience after placing an order.

A delivery time of this length may impact customer satisfaction, especially in competitive e-commerce markets where faster shipping is expected.

Further analysis could explore:

- impact of delivery time on repeat purchases

## Delivery Performance vs Estimated Time

In this step, we compare the actual delivery date with the estimated delivery date.

This allows us to evaluate whether the company is meeting its delivery promises.

```python
# convert estimated delivery date to datetime
orders['order_estimated_delivery_date'] = pd.to_datetime(orders['order_estimated_delivery_date'])

# calculate delay (actual - estimated)
orders['delivery_delay_days'] = (
    orders['order_delivered_customer_date'] - orders['order_estimated_delivery_date']
).dt.days

orders[['delivery_delay_days']].head()
```

```python
## Delivery Performance Segmentation

In this step, we classify orders into three categories:
- Early deliveries
- On-time deliveries
- Late deliveries

This helps us better understand overall delivery performance.
```

```python
def classify_delivery(x):
    if pd.isna(x):
        return 'Not Delivered'
    elif x < 0:
        return 'Early'
    elif x == 0:
        return 'On Time'
    else:
        return 'Late'

orders['delivery_status'] = orders['delivery_delay_days'].apply(classify_delivery)

orders['delivery_status'].value_counts()
```

## Business Insight - Delivery Performance Segmentation

The majority of orders (approximately 88,649) are delivered earlier than expected, indicating that delivery estimates are highly conservative and operations are generally efficient.

However, 6,535 orders are delivered late, representing a meaningful segment that may negatively impact customer satisfaction.

Additionally, 2,965 orders have not been delivered, which could indicate cancellations, logistics failures, or ongoing orders that require further investigation.

Only a small portion of orders are delivered exactly on time, suggesting that estimated delivery dates may not be optimized and could be adjusted to better reflect actual performance.

Overall, while early deliveries dominate, the presence of late and non-delivered orders highlights opportunities to improve reliability and customer experience.

## Merging Orders with Revenue Data

In this step, we merge the orders dataset with the order items dataset to associate each order with its corresponding revenue.

```python
df = orders.merge(order_items, on='order_id', how='left')

df.head()
```

## Revenue by Delivery Status

In this step, we calculate the total revenue associated with each delivery category.

This helps us understand how much money is affected by late deliveries.

```python
df.groupby('delivery_status')['price'].sum().sort_values(ascending=False)
```

## Business Insight - Revenue Impact of Late Deliveries

Late deliveries account for approximately 7–8% of total revenue, representing a significant portion of the business.

Although this segment is relatively small in volume, it involves nearly 1 million in revenue, making it financially meaningful.

These delayed orders are likely to have a disproportionate impact on customer satisfaction, retention, and brand perception.
